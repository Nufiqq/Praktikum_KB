### Import Package

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read Dataset

```python
df = pd.read_csv("titles.csv")
df
```

## 1. Grafik Batang

```python
x = df.genres[0:10]
y = df.tmdb_popularity[0:10]

plt.bar(x, y,)
plt.title("Perbandingan popularitas genre film")
plt.xticks(rotation=90)
plt.xlabel("Genre")
plt.ylabel("Popularitas")
plt.show()
```

Pada grafik batang di atas, dapat disimpulkan bahwa genre [romance, action dan drama] memiliki tingkat popularitas paling tinggi diantara 10 genre lainnya.
Adapun genre [documentation] memiliki popularitas paling sedikit.

## 2. Scatter Plot


```python
sns.scatterplot(data = df, x='type', y='release_year')
plt.title("Hubungan tipe film dengan tahun rilis")
plt.show()
```

Grafik diatas menunjukkan tipe film "SHOW" lebih banyak dirilis pada tahun 1980 keatas, sedangkan tipe film "MOVIE" mulai banyak dirilis pada tahun 1955 keatas

## 3. Grafik Pie

```python
labels = df['age_certification'].value_counts().rename_axis('age_certification').reset_index(name='Jumlah')
labels
```

```python
plt.figure(figsize=(8,8), dpi=300)
plt.title("Rating umur film")
plt.pie(df['age_certification'].value_counts(), labels=labels['age_certification'], autopct='%.1f%%', startangle=0)
plt.show()
```

Grafik pie diatas menampilkan persebaran rating-rating film yang dimana rating TV-MA memiliki film terbanyak dengan nilai 27.3% diantara 11 rating film lainnya. Dan yang meduduki penyebaran rating terndah yaitu NC-17 dengan nilai 0.5%

### 4. Grafik Heatmap

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.heatmap(df.corr(), cmap='Greens')
ax.set_title("Movies Dataset")
plt.show()
```

Grafik diatas menunjukkan korelasi antar variable yang menampilkan data dengan cara representasi warna yang berbeda-beda dan semakin tinggi angka suatu kelompok data maka warnanya akan semakin gelap.
