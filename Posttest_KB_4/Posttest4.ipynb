## 2. Data Cleaning


#### Menangani Nilai Null

```python
df.dtypes
```

```python
# Mengecek data null
df.isna().sum()
```

```python
imputer_string = SimpleImputer(strategy="most_frequent") #modus
```

```python
df['title'] = imputer_string.fit_transform(df[["title"]])
df['description'] = imputer_string.fit_transform(df[["description"]])
df['age_certification'] = imputer_string.fit_transform(df[["age_certification"]])
df['imdb_id'] = imputer_string.fit_transform(df[["imdb_id"]])
```

```python
imputer_float = SimpleImputer(strategy="mean")
```

```python
df['seasons'] = imputer_float.fit_transform(df[["seasons"]])
df['imdb_score'] = imputer_float.fit_transform(df[["imdb_score"]])
df['imdb_votes'] = imputer_float.fit_transform(df[["imdb_votes"]])
df['tmdb_popularity'] = imputer_float.fit_transform(df[["tmdb_popularity"]])
df['tmdb_score'] = imputer_float.fit_transform(df[["tmdb_score"]])
```

```python
df.isna().sum()
```
